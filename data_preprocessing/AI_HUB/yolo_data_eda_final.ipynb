```python
import pandas as pd
```

```python
df = pd.read_csv('output.csv')
```

```python
df
```

```python
import plotly.express as px

# í´ëž˜ìŠ¤ë³„ ë°ì´í„° ìˆ˜ ì§‘ê³„
class_counts = df['Class ID'].value_counts().reset_index()
class_counts.columns = ['Class ID', 'Frequency']

# í´ëž˜ìŠ¤ ë¶„í¬ë„ ì‹œê°í™”
fig = px.bar(class_counts, x='Class ID', y='Frequency', title='Train Class Distribution',
             labels={'Class ID': 'Class ID', 'Frequency': 'Frequency'},
             color='Frequency', color_continuous_scale='Viridis')

# ê·¸ëž˜í”„ ë³´ì—¬ì£¼ê¸°
fig.show()

fig.write_image('train_ class_distribution.png')
```

```python
class_counts
```

```python
df["BoundingBox Area"] = df["Width"] * df["Height"]
```

```python
df
```

```python
df['Image'].nunique()
```

```python
fig = px.box(df, x="Class ID", y="BoundingBox Area", title="Train-Class-wise Bounding Box Area Distribution")
fig.show()

fig.write_image('train_class_wise bounding.png')
```
